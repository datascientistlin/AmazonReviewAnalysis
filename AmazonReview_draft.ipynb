# Most Used Word Pairs 

To generate word pairs we did a similar processes as was done with the adjectives. Using NTLK, we parsed each review into an array of words. We performed used a feature from NTLK called "stemming" on each word in an attempt to get it's root. This was done to help join similar words such as "cancel" and "cancelling." With this new array we created a loop over all of the words in the review to get every two words into a set of pairs. Once we had the set of pairs we could perform a ratio of their usage for each product, group the pairs and sort by the sum of their ratios. This created the top 10 pairs as shown in the following table.
